# Most dangerous 10 diseases 

```python
# create a order series with o
# compiled a list of diseases in the number of deaths (from highest to lowest)
deseasesDeathsList = deathRecords.Icd10Code.value_counts()
deseasesDeathsList.head(10)
```

# Total Deaths occured in 2014 in US

```python
# calculate total deaths
totalDeaths = float(sum(deseasesDeathsList.values))
totalDeaths
```

```python
icdcodes = []
for x in deseasesDeathsList.index:
    tmp = icd10[icd10.Code == x]
    if tmp.empty:
        desc = ''
    else:
        desc = tmp.Description.values[0]
    icdcodes.append(desc)

s2 = pd.Series(icdcodes, index=deseasesDeathsList.index, name='Description')
print (s2.head(10))

s1 = pd.Series(deseasesDeathsList.values, index=deseasesDeathsList.index, name='Value')
s1.head(10)

```

```python

diseaseFatalityRank = pd.concat([s1, s2], axis=1)
diseaseFatalityRank.head(10)
```

```python
xx = [0]
for x in diseaseFatalityRank.Value / totalDeaths:
    xx.append(xx[-1] + x)
diseaseFatalityRank['AccumRelValue'] = xx[1:]
diseaseFatalityRank['Number'] = range(1,len(xx[1:])+1)
diseaseFatalityRank['Icd10Code'] = diseaseFatalityRank.index

diseaseFatalityRank.head(10)
```

```python
plt.figure(figsize=figsize)
title = 'Relative value of death'

for n in [9, 99, len(diseaseFatalityRank.AccumRelValue)-1]:
    title += "\n{:.0f}% of lives were caused by {} diseases  ".format( diseaseFatalityRank.AccumRelValue[n]*100.,n+1)

plt.title(title, fontsize=13)
plt.fill_between(range(diseaseFatalityRank.AccumRelValue.count()), diseaseFatalityRank.AccumRelValue, [0] * diseaseFatalityRank.AccumRelValue.count())
plt.xlabel('count of diseases')
plt.ylabel('relative value of total deaths')
plt.xscale('log')
plt.xlim(0,len(xx))
plt.ylim(0, 1)
plt.legend()
```

```python
# create new dataframe with part of general data
# leaving data when Age less than 120
# leaving data with DayOfWeekOfDeath less than 7
# leaving only 100 of the most dangerous diseases (for simplification)
df = deathRecords[deathRecords.Age < 120][deathRecords.DayOfWeekOfDeath < 8][deathRecords.Icd10Code.isin(deseasesDeathsList.iloc[0:100].index.tolist())]
```

```python
plt.figure(figsize=figsize)
plt.title('histogram of death vs age', fontsize=18)
_ = plt.hist(df.Age.tolist(), 60, alpha=0.9, label='M+F')
_ = plt.hist(df[df.Sex == 'M'].Age.tolist(), 60, alpha=0.5, label='M')
_ = plt.hist(df[df.Sex == 'F'].Age.tolist(), 60, alpha=0.5, label='F')
_ = plt.legend()
```

```python
d = []
for group in sorted(diseaseFatalityRank.Icd10Code.str[0].unique()):
    part = {}
    part['Group'] = group
    part['Value'] = diseaseFatalityRank[diseaseFatalityRank.Icd10Code.str[0] == group]['Value'].sum()
    d.append(part)
    
dfIcdGroup = pd.DataFrame(d).sort_index(by=['Value'])
dfIcdGroup['RelValue'] = dfIcdGroup.Value / totalDeaths

plt.figure(figsize=figsize)
plt.bar([x - 0.4 for x in range(dfIcdGroup.Value.count())], dfIcdGroup.RelValue)
_ = plt.xticks(range(len(dfIcdGroup.Value)), dfIcdGroup.Group)
plt.xlabel('Icd10Code group name')
plt.ylabel('Relative value of deaths')
plt.xlim(0, 23)

title = 'Most Dangerous Icd10Code Groups:'
for x in ['I', 'C', 'J']:
    title += "\n {} - {:.0f}%".format(x, dfIcdGroup[dfIcdGroup.Group == x].RelValue.values[0] * 100.)
plt.title(title, fontsize=18)
```
